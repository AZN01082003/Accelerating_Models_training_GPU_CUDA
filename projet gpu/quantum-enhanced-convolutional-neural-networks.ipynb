**Quantum-Enhanced Convolutional Neural Networks: A Comparative Analysis of Classical and Quantum Convolution for Image Classification**

This study investigates the implementation and performance of Quanvolutional Neural Networks (QNN), a novel approach combining quantum computing principles with classical convolutional neural networks for image classification tasks. We present a comparative analysis between classical and quantum convolution methods using the MNIST dataset. Our implementation utilizes a hybrid quantum-classical architecture, where quantum circuits process local image regions through parametrized rotations and measurements. The results demonstrate that the quantum approach achieves superior performance, with a best validation accuracy of 82.67% compared to 74.33% for the classical model. Furthermore, the quantum model exhibits higher confidence in correct predictions (90.09%) versus the classical model (76.87%). This study provides evidence for the potential advantages of quantum-enhanced neural networks in image processing tasks and suggests promising directions for future quantum machine learning applications.


The intersection of quantum computing and machine learning represents one of the most promising frontiers in computational science. As classical machine learning approaches face increasing challenges in processing complex data patterns, quantum computing offers potential advantages through its unique computational properties. Convolutional Neural Networks (CNNs) have demonstrated remarkable success in image processing tasks, yet they are limited by the classical nature of their operations.
Quanvolutional Neural Networks represent a novel paradigm that leverages quantum mechanical principles to enhance the feature extraction capabilities of traditional CNNs. By replacing classical convolution operations with quantum circuits, these networks can potentially capture more complex patterns and correlations within the input data. The quantum approach introduces non-linear transformations through quantum state evolution and measurement, potentially offering computational advantages over classical methods.
The fundamental principle behind quantum convolution involves processing local regions of input data through quantum circuits, where information is encoded into quantum states through parametrized rotations. This approach differs from classical convolution by utilizing quantum superposition and entanglement, potentially allowing for the extraction of features that would be computationally intensive or impossible to obtain through classical means.

**Experimental Setup**

Architecture Overview
The implemented system consists of two primary components:

- A quantum preprocessing layer that performs convolution operations using quantum circuits
- A classical neural network backend for classification

The quantum circuit utilizes 4 qubits and processes 2×2 input patches, mapping the results to multiple output channels. This architecture allows for direct comparison with classical convolution operations while maintaining quantum advantages.

Quantum Circuit Implementation
The quantum convolution process follows these steps:

a) Input Encoding:

Local image regions are encoded into quantum states using RY rotations
Input values are normalized and mapped to rotation angles in the range [0, π]
Four quantum features are extracted per patch:

- Mean intensity
- Maximum value
- Minimum value
- Standard deviation

b) Quantum Processing:

The quantum circuit applies a series of controlled-NOT (CNOT) operations between adjacent qubits
This creates entanglement between the qubits, enabling non-linear feature extraction
The circuit depth is optimized to balance computational complexity with feature extraction capability

c) Measurement:

Expectation values of PauliZ operators are measured for each qubit
These measurements form the output features for each processed patch

```python
def configure_tensorflow():
    try:
        tf.config.threading.set_intra_op_parallelism_threads(3)
        tf.config.threading.set_inter_op_parallelism_threads(2)
        
        tf.keras.mixed_precision.set_global_policy('float32')
        
    except Exception as e:
        print(f"Error during TensorFlow configuration: {e}")

configure_tensorflow()

n_epochs = 40
n_layers = 1
n_qubits = 4
n_train = 500
n_test = 300
batch_size = 16
CHUNK_SIZE = 8

SAVE_PATH = "./quanvolution/"
PREPROCESS = True
np.random.seed(42)
tf.random.set_seed(42)

mnist_dataset = keras.datasets.mnist
(train_images, train_labels), (test_images, test_labels) = mnist_dataset.load_data()

train_images = np.array(train_images[:n_train]) / 255.0
train_labels = np.array(train_labels[:n_train])
test_images = np.array(test_images[:n_test]) / 255.0
test_labels = np.array(test_labels[:n_test])

train_images = tf.convert_to_tensor(train_images[..., np.newaxis], dtype=tf.float32)
test_images = tf.convert_to_tensor(test_images[..., np.newaxis], dtype=tf.float32)

def clear_memory():
    gc.collect()
    tf.keras.backend.clear_session()

def get_optimal_thread_count():
    try:
        physical_cores = 2
        logical_cores = 4
        available_memory = psutil.virtual_memory().available / (1024 ** 3)
        
        optimal_workers = min(
            3,
            int(available_memory / 2),
            logical_cores - 1
        )
        
        return max(2, optimal_workers)
    except Exception as e:
        print(f"Error during optimal thread calculation: {e}")
        return 2

dev = qml.device("default.qubit", wires=n_qubits)

@qml.qnode(dev, interface="tf", diff_method="parameter-shift")
def circuit(inputs):
    inputs = np.nan_to_num(inputs, nan=0.0, posinf=1.0, neginf=0.0)
    inputs = np.clip(inputs, 0, 1)

    for i in range(n_qubits):
        qml.RY(inputs[i] * np.pi, wires=i)
    
    for i in range(n_qubits - 1):
        qml.CNOT(wires=[i, i + 1])
    
    return [qml.expval(qml.PauliZ(i)) for i in range(n_qubits)]
    

def process_batch(images_batch):
    return [process_image(image) for image in images_batch]

def process_single_patch(patch):
    try:
        patch_np = patch.numpy() if isinstance(patch, tf.Tensor) else patch
        patch_np = patch_np.astype(np.float32)
        
        patch_np = (patch_np - np.min(patch_np)) / (np.max(patch_np) - np.min(patch_np) + 1e-8)
        
        with qml.tape.QuantumTape() as tape:
            result = circuit(patch_np)
        
        return np.array(result, dtype=np.float32)
    except Exception as e:
        print(f"Error in process_single_patch: {e}")
        return np.zeros(n_qubits, dtype=np.float32)

def process_patch(patch):
    return tf.py_function(
        func=process_single_patch,
        inp=[patch],
        Tout=tf.float32
    )

@tf.function
def update_output_slice(output, j, k, result):
    for i in range(n_qubits):
        output = tf.tensor_scatter_nd_update(
            output,
            [[j, k, i]],
            [result[i]]
        )
    return output

@tf.function(experimental_relax_shapes=True)
def process_image(image):
    output = tf.zeros((7, 7, n_qubits), dtype=tf.float32)
    
    def calculate_std(x):
        mean = tf.reduce_mean(x)
        squared_diff = tf.square(x - mean)
        variance = tf.reduce_mean(squared_diff)
        return tf.sqrt(variance + 1e-8)
    
    for j in range(0, 7):
        for k in range(0, 7):
            j_start = j * 4
            k_start = k * 4
            
            patch_region = image[j_start:j_start+4, k_start:k_start+4, 0]
            
            patch = tf.stack([
                tf.reduce_mean(patch_region),
                tf.reduce_max(patch_region),
                tf.reduce_min(patch_region),
                calculate_std(patch_region)
            ])
            
            result = process_patch(patch)
            output = tf.tensor_scatter_nd_update(
                output,
                [[j, k, i] for i in range(n_qubits)],
                result
            )
    
    return output

def preprocess_quantum(images):
    processed_images = []
    optimal_workers = get_optimal_thread_count()
    
    print(f"Using {optimal_workers} workers for preprocessing")
    
    batch_size = 20
    for i in tqdm(range(0, len(images), batch_size)):
        batch = images[i:i + batch_size]
        
        with ThreadPoolExecutor(max_workers=optimal_workers) as executor:
            futures = [executor.submit(process_image, img) for img in batch]
            
            for future in futures:
                try:
                    result = future.result()
                    processed_images.append(result)
                except Exception as e:
                    print(f"Error processing an image: {e}")
                    processed_images.append(tf.zeros((7, 7, n_qubits), dtype=tf.float32))
        
        gc.collect()
    
    return tf.stack(processed_images)

if PREPROCESS:
    print("Preprocessing training images...")
    q_train_images = preprocess_quantum(train_images)
    
    print("Preprocessing test images...")
    q_test_images = preprocess_quantum(test_images)

    train_images_resized = resize_classical_images(train_images)
    test_images_resized = resize_classical_images(test_images)
    
    os.makedirs(SAVE_PATH, exist_ok=True)
    
    np.save(SAVE_PATH + "q_train_images.npy", q_train_images.numpy())
    np.save(SAVE_PATH + "q_test_images.npy", q_test_images.numpy())
    print("Preprocessing completed and results saved.")

    print("\nVisualizing samples...")
    visualize_preprocessing(
        train_images[:5], 
        q_train_images[:5],
        train_images_resized[:5]
    )   

    print("\nValue distribution...")
    plot_value_distributions(train_images[:5], q_train_images[:5])

    print("\nAnalyzing quantum feature importance...")
    analyze_quantum_feature_importance()

    clear_memory()

else:
    q_train_images = tf.convert_to_tensor(np.load(SAVE_PATH + "q_train_images.npy"))
    q_test_images = tf.convert_to_tensor(np.load(SAVE_PATH + "q_test_images.npy"))

def create_base_model(input_shape):
    model = keras.Sequential([
        keras.layers.Input(shape=input_shape),
        keras.layers.Flatten(),
        keras.layers.Dense(128, activation='relu'),
        keras.layers.Dropout(0.2),
        keras.layers.Dense(10, activation='softmax')
    ])
    
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    
    return model

train_images_resized = resize_classical_images(train_images)
test_images_resized = resize_classical_images(test_images)

quantum_model = create_base_model(input_shape=(7, 7, n_qubits))
classical_model = create_base_model(input_shape=(7, 7, 1))


print("Training quantum model...")
import time
start=time.time()
q_history = quantum_model.fit(
    q_train_images, train_labels,
    validation_data=(q_test_images, test_labels),
    batch_size=32,
    epochs=n_epochs,
    verbose=1
)
end= time.time()
quantum_training_time = end - start
print(f"Quantum model training completed in {quantum_training_time:.2f} seconds.")


print("Training classical model...")
start_time=time.time()
c_history = classical_model.fit(
    train_images_resized, train_labels,
    validation_data=(test_images_resized, test_labels),
    batch_size=32,
    epochs=n_epochs,
    verbose=1
)

end_time = time.time()
classical_training_time = end_time - start_time
print(f"Classical model training completed in {classical_training_time:.2f} seconds.")
```

```python


import time
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pennylane as qml
import gc
import psutil
from concurrent.futures import ThreadPoolExecutor
from tqdm import tqdm

# Configuration TensorFlow
def configure_tensorflow():
    tf.config.threading.set_intra_op_parallelism_threads(3)
    tf.config.threading.set_inter_op_parallelism_threads(2)
    tf.keras.mixed_precision.set_global_policy('float32')

configure_tensorflow()

# Paramètres globaux
n_epochs = 20
n_layers = 1
n_qubits = 4
n_train = 300
n_test = 100
batch_size = 1000
CHUNK_SIZE = 60     #8
SAVE_PATH = "./quanvolution/"
PREPROCESS = True
np.random.seed(42)
tf.random.set_seed(42)


def clear_memory():
    gc.collect()
    tf.keras.backend.clear_session()

def get_optimal_thread_count():
    physical_cores = 2
    logical_cores = 4
    available_memory = psutil.virtual_memory().available / (1024 ** 3)
    optimal_workers = min(3, int(available_memory / 2), logical_cores - 1)
    return max(2, optimal_workers)

dev = qml.device("default.qubit", wires=n_qubits)

@qml.qnode(dev, interface="tf", diff_method="parameter-shift")
def circuit(inputs):
    inputs = np.clip(inputs, 0, 1)
    for i in range(n_qubits):
        qml.RY(inputs[i] * np.pi, wires=i)
    for i in range(n_qubits - 1):
        qml.CNOT(wires=[i, i + 1])
    return [qml.expval(qml.PauliZ(i)) for i in range(n_qubits)]

def process_single_patch(patch):
    patch_np = patch.numpy() if isinstance(patch, tf.Tensor) else patch
    patch_np = (patch_np - np.min(patch_np)) / (np.max(patch_np) - np.min(patch_np) + 1e-8)
    return np.array(circuit(patch_np), dtype=np.float32)

def process_patch(patch):
    return tf.py_function(func=process_single_patch, inp=[patch], Tout=tf.float32)

def process_image(image):
    output = tf.zeros((7, 7, n_qubits), dtype=tf.float32)
    for j in range(7):
        for k in range(7):
            patch_region = image[j*4:j*4+4, k*4:k*4+4, 0]
            patch = tf.stack([
                tf.reduce_mean(patch_region),
                tf.reduce_max(patch_region),
                tf.reduce_min(patch_region),
                tf.math.reduce_std(patch_region)
            ])
            result = process_patch(patch)
            output = tf.tensor_scatter_nd_update(output, [[j, k, i] for i in range(n_qubits)], result)
    return output

def preprocess_quantum(images):
    processed_images = []
    batch_size = 200
    for i in tqdm(range(0, len(images), batch_size)):
        batch = images[i:i + batch_size]
        with ThreadPoolExecutor(max_workers=get_optimal_thread_count()) as executor:
            results = list(executor.map(process_image, batch))
        processed_images.extend(results)
    return tf.stack(processed_images)

# Mesure du temps de prétraitement
start_preprocessing = time.time()
q_train_images = preprocess_quantum(train_images)
q_test_images = preprocess_quantum(test_images)
end_preprocessing = time.time()
preprocessing_time = end_preprocessing - start_preprocessing

# Définition du modèle

def create_base_model(input_shape):
    model = keras.Sequential([
        keras.layers.Input(shape=input_shape),
        keras.layers.Flatten(),
        keras.layers.Dense(512, activation='relu'),
        keras.layers.BatchNormalization(),
        keras.layers.Dropout(0.3),
        keras.layers.Dense(256, activation='relu'),
        keras.layers.BatchNormalization(),
        keras.layers.Dropout(0.3),
        keras.layers.Dense(128, activation='relu'),
        keras.layers.Dropout(0.2),
        keras.layers.Dense(64, activation='relu'),
        keras.layers.Dense(10, activation='softmax')
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model

quantum_model = create_base_model(input_shape=(7, 7, n_qubits))
classical_model = create_base_model(input_shape=(28, 28, 1))

# Mesure du temps d'entraînement du modèle quantique
start_training_quantum = time.time()
q_history = quantum_model.fit(q_train_images, train_labels, validation_data=(q_test_images, test_labels),
                              batch_size=32, epochs=n_epochs, verbose=1)
end_training_quantum = time.time()
quantum_training_time = end_training_quantum - start_training_quantum

# Mesure du temps d'entraînement du modèle classique
start_training_classical = time.time()
c_history = classical_model.fit(train_images, train_labels, validation_data=(test_images, test_labels),
                                batch_size=32, epochs=n_epochs, verbose=1)
end_training_classical = time.time()
classical_training_time = end_training_classical - start_training_classical

# Rapport final des temps
print(f"Temps de prétraitement quantique : {preprocessing_time:.2f} secondes")
print(f"Temps d'entraînement du modèle quantique : {quantum_training_time:.2f} secondes")
print(f"Temps d'entraînement du modèle classique : {classical_training_time:.2f} secondes")
```

```python
import time
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pennylane as qml
gc.collect()
tf.keras.mixed_precision.set_global_policy('float32')

# Paramètres globaux
n_epochs = 100
n_layers = 1
n_qubits = 4
n_train = 1000
n_test = 300
batch_size = 16
CHUNK_SIZE = 8
SAVE_PATH = "./quanvolution/"
PREPROCESS = True
np.random.seed(42)
tf.random.set_seed(42)

# Chargement des données MNIST
mnist_dataset = keras.datasets.mnist
(train_images, train_labels), (test_images, test_labels) = mnist_dataset.load_data()

train_images = np.array(train_images[:n_train]) / 255.0
train_labels = np.array(train_labels[:n_train])
test_images = np.array(test_images[:n_test]) / 255.0
test_labels = np.array(test_labels[:n_test])

train_images = tf.convert_to_tensor(train_images[..., np.newaxis], dtype=tf.float32)
test_images = tf.convert_to_tensor(test_images[..., np.newaxis], dtype=tf.float32)

def clear_memory():
    gc.collect()
    tf.keras.backend.clear_session()

dev = qml.device("default.qubit", wires=n_qubits)

@qml.qnode(dev, interface="tf", diff_method="parameter-shift")
def circuit(inputs):
    inputs = tf.clip_by_value(inputs, 0, 1)
    for i in range(n_qubits):
        qml.RY(inputs[i] * np.pi, wires=i)
    qml.broadcast(qml.CNOT, wires=range(n_qubits), pattern="ring")
    return [qml.expval(qml.PauliZ(i)) for i in range(n_qubits)]

def process_single_patch(patch):
    patch = tf.reshape(patch, [-1])
    patch = tf.clip_by_value(patch, 0, 1)
    return circuit(patch)

def process_image(image):
    patches = tf.image.extract_patches(images=tf.expand_dims(image, axis=0),
                                       sizes=[1, 4, 4, 1],
                                       strides=[1, 4, 4, 1],
                                       rates=[1, 1, 1, 1],
                                       padding='VALID')
    patches = tf.reshape(patches, [-1, 16])
    results = tf.map_fn(process_single_patch, patches, dtype=tf.float32)
    return tf.reshape(results, (7, 7, n_qubits))

def preprocess_quantum(images):
    return tf.map_fn(process_image, images, dtype=tf.float32)

# Mesure du temps de prétraitement
start_preprocessing = time.time()
q_train_images = preprocess_quantum(train_images)
q_test_images = preprocess_quantum(test_images)
end_preprocessing = time.time()
preprocessing_time = end_preprocessing - start_preprocessing

# Définition du modèle
def create_base_model(input_shape):
    model = keras.Sequential([
        keras.layers.Input(shape=input_shape),
        keras.layers.Flatten(),
        keras.layers.Dense(512, activation='relu'),
        keras.layers.BatchNormalization(),
        keras.layers.Dropout(0.3),
        keras.layers.Dense(256, activation='relu'),
        keras.layers.BatchNormalization(),
        keras.layers.Dropout(0.3),
        keras.layers.Dense(128, activation='relu'),
        keras.layers.Dropout(0.2),
        keras.layers.Dense(64, activation='relu'),
        keras.layers.Dense(10, activation='softmax')
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model

quantum_model = create_base_model(input_shape=(7, 7, n_qubits))
classical_model = create_base_model(input_shape=(28, 28, 1))

# Mesure du temps d'entraînement du modèle quantique
start_training_quantum = time.time()
q_history = quantum_model.fit(q_train_images, train_labels, validation_data=(q_test_images, test_labels),
                              batch_size=32, epochs=n_epochs, verbose=1)
end_training_quantum = time.time()
quantum_training_time = end_training_quantum - start_training_quantum

# Mesure du temps d'entraînement du modèle classique
start_training_classical = time.time()
c_history = classical_model.fit(train_images, train_labels, validation_data=(test_images, test_labels),
                                batch_size=32, epochs=n_epochs, verbose=1)
end_training_classical = time.time()
classical_training_time = end_training_classical - start_training_classical

# Rapport final des temps
print(f"Temps de prétraitement quantique : {preprocessing_time:.2f} secondes")
print(f"Temps d'entraînement du modèle quantique : {quantum_training_time:.2f} secondes")
print(f"Temps d'entraînement du modèle classique : {classical_training_time:.2f} secondes")
```

```python
import pennylane as qml
import tensorflow as tf
import numpy as np

# Configuration du simulateur GPU
n_qubits = 4
dev = qml.device('lightning.gpu', wires=n_qubits)

# Vérification de l'utilisation du GPU par TensorFlow
physical_devices = tf.config.list_physical_devices('GPU')
if physical_devices:
    tf.config.experimental.set_memory_growth(physical_devices[0], True)
    tf.config.optimizer.set_jit(True)
    print("GPU détecté et optimisé.")

# Définition du circuit quantique optimisé
@qml.qnode(dev, interface="tf", diff_method="adjoint")
def quantum_circuit(inputs):
    qml.templates.AngleEmbedding(inputs, wires=range(n_qubits))
    qml.BasicEntanglerLayers(weights=np.random.random((3, n_qubits)), wires=range(n_qubits))
    return [qml.expval(qml.PauliZ(i)) for i in range(n_qubits)]

# Fonction de traitement par lot avec parallélisation
batch_size = 32

def process_batch(batch):
    batch = tf.cast(batch, tf.float32)
    results = []
    for img in batch:
        result = quantum_circuit(img[:n_qubits])
        results.append(result)
    return tf.convert_to_tensor(results)

# Simulation de données
train_images = np.random.random((1000, n_qubits)).astype(np.float32)

# Chargement et parallélisation des données avec tf.data.Dataset
dataset = tf.data.Dataset.from_tensor_slices(train_images)
dataset = (dataset.batch(batch_size)
                .map(lambda x: tf.py_function(func=process_batch, inp=[x], Tout=tf.float32),
                     num_parallel_calls=tf.data.AUTOTUNE)
                .prefetch(buffer_size=tf.data.AUTOTUNE))

# Optimisation de la compilation JIT avec XLA pour TensorFlow
def loss_fn(predictions):
    return tf.reduce_mean(predictions)

optimizer = tf.keras.optimizers.Adam()

@tf.function(jit_compile=True)
def train_step(batch):
    with tf.GradientTape() as tape:
        predictions = process_batch(batch)
        loss = loss_fn(predictions)
    gradients = tape.gradient(loss, quantum_circuit.trainable_variables)
    optimizer.apply_gradients(zip(gradients, quantum_circuit.trainable_variables))
    return loss

# Entraînement
for epoch in range(5):
    for batch in dataset:
        loss = train_step(batch)
    print(f"Epoch {epoch + 1}, Loss: {loss.numpy()}")
```
